```python
import pandas as pd
```

```python
date = input("Input date ex. 20190401")
hour = input("Input hour ex. 100000")
```

```python
blacklist = pd.read_csv('/Users/rahmaahmed/Documents/WikipediaPageCounts/blacklist_domains_and_pages', sep=' ', header=None)
blacklist.columns = ['domain_code', 'page_title']
print(blacklist)
```

```python
data = pd.read_csv(f"/Users/rahmaahmed/Documents/WikipediaPageCounts/pageviews-{date}-{hour}.csv", sep=' ', header=None)
data.columns = ['domain_code', 'page_title', 'count_views', 'total_response_size']
print(data)
```

```python
keys = list(blacklist.columns.values)
i1 = data.set_index(keys).index
i2 = blacklist.set_index(keys).index
non_blacklisted_data = data[~i1.isin(i2)]
print(non_blacklisted_data)
```

```python
unique_domain_code = non_blacklisted_data['domain_code'].nunique()
print(unique_domain_code)
```

```python
blacklisted_data = data[i1.isin(i2)]
print(blacklisted_data)
```

```python
final_df = non_blacklisted_data.groupby(['domain_code', 'page_title'])['count_views'].agg('sum').reset_index()
final_df.columns = ['domain_code', 'page_title', 'count_views']

df1 = final_df.groupby("domain_code", group_keys=False).apply(lambda g: g.nlargest(25, "count_views"))
df2 = pd.DataFrame(df1)
df2.to_csv(f"top25-{date}-{hour}",sep='\t')
```
